```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel("537Longevity.xlsx")
```

```python
df = df.fillna(0)
```

```python
df.columns
```

```python
new_df = pd.DataFrame(columns = ('Cage ID', 'Sex','Virgin','Yeast', 'Density', 'Total', 'Day','Event'))
time_list = np.array(df.columns[7:])
time_list = np.array([time[4:] for time in time_list])
time_list = time_list.astype(int)
for i in range(len(df)):
    row = df.iloc[i]
    vial_id = row.iloc[5]
    sex = row.iloc[3]
    virgin = row.iloc[2]
    yeast = row.iloc[1]
    density = row.iloc[4]
    total = row.iloc[6]
    n_deaths_list = np.array(row.iloc[7:])
    for t in range(len(time_list)):
        n_death = int(n_deaths_list[t])
        if n_death != 0:
            new_df_rows = {'Cage ID':vial_id, 'Sex':sex, 'Virgin':virgin, 'Yeast':yeast, 'Density':density, 'Total':total, 
                           'Day': np.repeat(time_list[t], n_death), 'Event': np.ones(n_death)}
            new_df = pd.concat([new_df,pd.DataFrame(new_df_rows)])
```

```python
new_df.to_csv('537 Longevity sorted data.csv', index=False)
```
